Based on the research output, I replaced the abbreviations of the race/ethnicity categories with the whole phrases the abbreviations stand for.
The research output can be found here: https://dataverse.harvard.edu/file.xhtml?persistentId=doi:10.7910/DVN/KERZQY/SN8AMT&version=3.0

```python
repl = {'w': 'White', 'b': 'Black', 'a': 'Asian', 'l': 'Latino', 'i': 'Indigenous', 's': 'South Asian', 'n': 'Near Eastern'}
racial_lines_df = racial_lines_df.replace({'CHARACTER_RACE': repl})
racial_lines_df['CHARACTER_RACE'].value_counts()
```

```python
# I wanted to see the most important pieces of information.
racial_lines_df[['CHARACTER_NAME', 'ACTOR_NAME', 'GENDER', 'CHARACTER_RACE', 'AGE', 'EXTRACTED_WORDS']].head(20)
```

Research Question 1: What is the ratio/distribution of words uttered by non-white characters belonging to different races/ethnicities?

```python
# I created a dataframe of non-white characters.
non_white_characters = racial_lines_df['CHARACTER_RACE'] != 'White'
non_white_characters = racial_lines_df[non_white_characters]
non_white_characters
```

```python
# I wanted to see how many words are uttered per race/ethnicity in a descending order.
non_white_words = non_white_characters.groupby('CHARACTER_RACE')['EXTRACTED_WORDS'].sum().sort_values(ascending = False)
non_white_words
```

```python
# I plotted these results in the form of a pie chart with percentages.
non_white_words.plot(kind='pie', figsize =(5,5), autopct='%1.1f%%', title = 'The number of words uttered by non-white characters according to their race/ethnicity');
```

```python
# I wanted to experiment with another type of plot as well.
non_white_words.plot(kind='bar', xlabel='character race', ylabel='number of words', title = 'The number of words uttered by non-white characters according to their race/ethnicity', color = 'brown');
```

Research Question 2: What kind of names are given to Latino characters? Are Latino characters given stereotypical Latino names, e.g. Juan, Carlos, Miguel?

```python
# I created a 'Latino' filter. Then, I selected some columns that interested me, and sorted the values in a descending order of the age of the character, and looked at the 50 oldest characters.
latino_filter = racial_lines_df['CHARACTER_RACE'] == 'Latino'
latino_filter = racial_lines_df[latino_filter]
latino_filter[['CHARACTER_NAME', 'ACTOR_NAME', 'GENDER', 'CHARACTER_RACE', 'AGE', 'EXTRACTED_WORDS']].sort_values(by='AGE', ascending= False).head(50)
```

Research Question 3: How is the number of non-white characters distributed along the lines of gender? How many male and female characters are there of each race/ethnicity?

```python
# I created a dataframe of non-white male characters.
non_white_male_characters = non_white_characters['GENDER'] == 'm'
non_white_male_characters = non_white_characters[non_white_male_characters]
non_white_male_characters
```

```python
# just checking if there are any white characters
non_white_male_characters[non_white_male_characters['CHARACTER_RACE'] == 'White']
```

```python
# just checking if there are any female characters
non_white_male_characters[non_white_male_characters['CHARACTER_RACE'] == 'f']
```

```python
# I created a dataframe of non-white female characters.
non_white_female_characters = non_white_characters['GENDER'] == 'f'
non_white_female_characters = non_white_characters[non_white_female_characters]
non_white_female_characters
```

```python
# just checking if there are any white characters
non_white_female_characters[non_white_female_characters['CHARACTER_RACE'] == 'White']
```

```python
# just checking if there are any male characters
non_white_female_characters[non_white_female_characters['CHARACTER_RACE'] == 'm']
```

```python
# How many female characters are there of each race/ethnicity?
fdf = non_white_female_characters.groupby('CHARACTER_RACE')['CHARACTER_ID'].count()
fdf = pd.DataFrame(fdf)
fdf = fdf.rename(columns={'CHARACTER_ID': 'number of female characters'})
fdf
```

```python
# How many male characters are there of each race/ethnicity?
mdf = non_white_male_characters.groupby('CHARACTER_RACE')['CHARACTER_ID'].count()
mdf = pd.DataFrame(mdf)
mdf = mdf.rename(columns={'CHARACTER_ID': 'number of male characters'})
mdf
```

```python
# Let's visualize these numbers in the form of a line plot.
ax = fdf.plot(kind='line', xlabel='character race', ylabel='number of characters', color='orange')
mdf.plot(ax=ax, kind='line', color='green');
plt.tight_layout()
ax.figure.savefig('non-white male and female characters.png')
# For some reason the category South Asian is missing from the plot.
# It may be due to the fact that there are no Near Eastern female characters and that created a 'break' in the line plot(?).
```
